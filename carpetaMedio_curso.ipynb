Parte 1: Carga y Limpieza (Pandas y NumPy)
Importa las librerías necesarias (pandas, numpy, matplotlib.pyplot).
Lee el archivo CSV desde la URL proporcionada y guárdalo en un DataFrame.
Imprime las primeras 5 filas y usa un comando para mostrar los tipos de datos de cada columna.
Identifica si hay valores nulos en la columna horsepower (caballos de fuerza). Reemplaza esos valores nulos con la mediana de esa misma columna usando NumPy y Pandas.
Crea una nueva columna llamada relacion_peso_potencia que sea el resultado de dividir la columna weight (peso) entre horsepower.

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# URL del dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"

# Leer el archivo CSV y guardarlo en un DataFrame
df = pd.read_csv(url)

# Mostrar las primeras 5 filas
print("Primeras 5 filas del dataset:")
print(df.head())

# Mostrar los tipos de datos de cada columna
print("\nTipos de datos de las columnas:")
print(df.dtypes)

# Verificar valores nulos en la columna horsepower
print("\nValores nulos en horsepower:")
print(df['horsepower'].isnull().sum())

# Calcular la mediana de horsepower
mediana_hp = df['horsepower'].median()

# Reemplazar valores nulos con la mediana
df['horsepower'] = df['horsepower'].fillna(mediana_hp)

# Crear nueva columna relacion_peso_potencia
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']

# Mostrar las primeras filas con la nueva columna
print("\nDataset con la nueva columna relacion_peso_potencia:")
print(df.head())
```

Parte 2: Visualización de Datos (Matplotlib) Genera las siguientes gráficas. Todas deben incluir título y etiquetas en los ejes X y Y.
Histograma: Muestra la distribución de la aceleración (acceleration) de los vehículos.
Gráfico de Dispersión (Scatter): Grafica el Desplazamiento del motor (displacement) en el eje X versus los Caballos de fuerza (horsepower) en el eje Y.
Gráfico de Barras: Muestra el promedio de rendimiento (mpg - millas por galón) agrupado por el país de origen (origin). Pista: Tendrás que agrupar los datos primero con Pandas.

```python
# Histograma de acceleration
plt.hist(df['acceleration'])
plt.title("Acceleration")
plt.xlabel("Acceleration")
plt.ylabel("Frecuencia")
plt.show()


# Scatter plot displacement vs horsepower
plt.scatter(df['displacement'], df['horsepower'])
plt.title("Displacement vs Horsepower")
plt.xlabel("Displacement")
plt.ylabel("Horsepower")
plt.show()


# Promedio de mpg por origen
promedio = df.groupby('origin')['mpg'].mean()

plt.bar(promedio.index, promedio.values)
plt.title("MPG por origen")
plt.xlabel("Origin")
plt.ylabel("MPG promedio")
plt.show()
```
